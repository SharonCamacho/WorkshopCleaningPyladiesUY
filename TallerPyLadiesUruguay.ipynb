### **Visualización**

En esta etapa para agilizar el análisis vamos a utlilizar solo 7 columnas.

Por esto, el primer paso es filtrar el dataset con una lista de las variables que utilizaremos

````
df = df[['Age', 'JobInvolvement', 'MaritalStatus', 'MonthlyIncome', 'OverTime', 'YearsSinceLastPromotion', 'Attrition']]
````

Tenemos entonces 3 variables categóricas y 4 numéricas

````
df.dtypes.value_counts()
````

Estas son las librerías con las que trabajaremos

````
import seaborn as sns
import matplotlib.pyplot as plt
````

#### Catplot

**Desgaste según año de último ascenso**

````
sns.catplot(data=df, x="YearsSinceLastPromotion",  kind="count")
````

````
sns.catplot(x="YearsSinceLastPromotion", data=df,  kind="count", hue='Attrition')
````

````
sns.catplot(x="YearsSinceLastPromotion", data=df,  kind="count", hue='Attrition')
plt.yscale('log')
````

**Desgaste según Participación Laboral**

````
sns.catplot(x="JobInvolvement", data=df,  kind="count", hue='Attrition')
plt.yscale('log')
````

#### BoxPlot

El boxplot es u tipo de catplot

````
g1 = sns.catplot(x="MonthlyIncome", y="Attrition", 
                kind="box", orient="h", height=1.5, aspect=4,
                data=df)
````

````
g = sns.catplot(x="MonthlyIncome", y="Attrition", row='OverTime',
                kind="box", orient="h", height=1.5, aspect=4,
                data=df)
````

#### Displot

````
sns.displot(df, x='MonthlyIncome')

plt.xlabel('Ingresos Mesuales')
plt.ylabel('# Registros')
plt.title('Ingresos Mensuales')
````

#### Correlaciones

````
sns.pairplot(df, hue='Attrition')
````

````
correlation_mat = df.corr()
#plt.figure(figsize=(15, 10))
sns.heatmap(correlation_mat, annot = True, cmap='coolwarm')
plt.show()
````
